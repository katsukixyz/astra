```python
from shapely.geometry import Point
points = []
min_x, min_y, max_x, max_y = maps.total_bounds
```

```python
for i in range(10000):
    x = random.uniform(min_x, max_x)
    y = random.uniform(min_y, max_y)
    point = Point(x, y)
    if(maps.contains(point).item()):
        print(i)
        points.append((x,y))
```
